```python
# drop postal code
df=df.drop(['Postal Code'],axis=1)
```

```python
plt.figure(figsize=(20,12))
df.hist(bins=50)
plt.show()
```

Observation: <br>
1. Most customers tends to buy quantity of 2 and 3. <br>
2. Discount give maximum 0 to 20 percent.

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar()
plt.show()
```

Profit and Sales are maximum in Consumer Segment and minimum in Home Office Segment.

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar()
plt.show()
```

Profit is maximum in west region and minimum in south region.

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar()
plt.show()
```

Profit is maximum in Technolgy and Office Supplies and minimum in Furniture.
